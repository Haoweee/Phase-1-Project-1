## Data Cleaning

```python
df.info()
```

### Dropping Unnessary Columns

```python
#Drop columns that have high missingness and no pertinence to end analysis
drop_columns = [
    'latitude', 
    'longitude', 
    'airport_code', 
    'airport_name',  
    'schedule', 
    'air_carrier',
    'publication_date'
]

df = df.drop(drop_columns, axis=1)

# accident number, event_id, and registration_number :  no way to compare them if they are all distinct 
```

### Filling in NaN values for Categorical Columns

```python
def replace_categorical_values(dataframe, column_list, replacement_text):
    for column in column_list:
        #Fill NaN columns with the replacement text
        dataframe[column].fillna(replacement_text, inplace=True)
    return dataframe.info()
```

```python
# Replace NaN values in categorical columns

column_list = [
    'location', 
    'country', 
    'injury_severity', 
    'aircraft_damage', 
    'aircraft_category', 
    'registration_number', 
    'make', 
    'model',
    'amateur_built',
    'engine_type',
    'far_description',
    'purpose_of_flight',
    'weather_condition',
    'broad_phase_of_flight',
    'report_status'  
] 

replace_categorical_values(df, column_list, 'unknown')
```

### Filling in NaN values for Numerical Columns

For numerical columns we will need to determine whether to use the mean, median, or mode to impute the missing values. 

Below are histograms to help us see the shape of the distribution in each column.

```python
# Create subplots with 2 rows and 2 columns
fig, axes = plt.subplots(3, 2, figsize=(12, 10))

# Flatten the axes array for easy indexing
axes = axes.flatten()

# Iterate over each column and plot a histogram
columns = ['number_of_engines', 'total_fatal_injuries', 'total_serious_injuries', 'total_minor_injuries', 'total_uninjured']
for i, column in enumerate(columns):
    # Drop NaN values before plotting
    data = df[column].dropna()
    
    # Plot histogram
    axes[i].hist(data, bins=300, color='skyblue', edgecolor='black')
    
    # Set title for each subplot
    axes[i].set_title(f'Histogram of {column}')
    
    # Set labels for x and y axes
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frequency')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plot
plt.show()
```

- All of the histograms show a strong right skew. 
- We will use the median value to impute missing values.

```python
def replace_numerical_values(dataframe, column_list, central_tendency):
    for column in column_list:
        # Calculate the central tendency value for the column
        central_value = central_tendency(dataframe[column].dropna())
        # Fill NaN values with the calculated central tendency value
        dataframe[column].fillna(central_value, inplace=True)
    return dataframe.info()
```

```python
#Fill the NaN values listed in numerical columns with the columns median value
numerical_columns = [
    'number_of_engines', 
    'total_fatal_injuries', 
    'total_serious_injuries', 
    'total_minor_injuries', 
    'total_uninjured'
]

replace_numerical_values(df, numerical_columns, np.median)
```

```python
#With Risk defined as total injuries, we will create a new column called injuries 
df['total_injuries'] = df['total_fatal_injuries'] + df['total_serious_injuries'] + df['total_minor_injuries']
```
